# Resuelva

$ x_1 + x_2 = 6$

$  -2 x_1 +  x_2 = 0$

Método de Gauss-Jacobi

```python
def gauss_jacobi(A, b, x0, tol=1e-5, max_iter=1000):

    n = len(A)
    x = x0[:]
    x_new = [0] * n
    trayectoria = [x0[:]]

    for it in range(max_iter):
        for i in range(n):
            suma1 = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - suma1) / A[i][i]

        trayectoria.append(x_new[:])

        if all(abs(x_new[i] - x[i]) < tol for i in range(n)):
            return x_new, trayectoria

        x[:] = x_new

    raise ValueError("No convergió en el número máximo de iteraciones")

A = [[1, 1], [-2, 1]]
b = [6, 0]
x0 = [0, 0]

sol_jacobi, trayectoria_jacobi = gauss_jacobi(A, b, x0, tol=1e-5, max_iter=1000)

print("Solución:", sol_jacobi)
print("Trayectoria:")
for i, x in enumerate(trayectoria_jacobi):
    print(f"Iteración {i}: {x}")
```

Método de Gauss-Seidel


```python
def gauss_seidel(A, b, x0, tol=1e-5, max_iter=1000):
    n = len(A)
    x = x0[:]
    trayectoria = [x0[:]]

    for it in range(max_iter):
        x_new = x[:]

        for i in range(n):
            suma1 = sum(A[i][j] * x_new[j] for j in range(i))
            suma2 = sum(A[i][j] * x[j] for j in range(i + 1, n))
            x_new[i] = (b[i] - suma1 - suma2) / A[i][i]

        trayectoria.append(x_new[:])

        if all(abs(x_new[i] - x[i]) < tol for i in range(n)):
            return x_new, trayectoria

        x[:] = x_new

    raise ValueError("No convergió en el número máximo de iteraciones")

A = [[1, 1], [-2, 1]]
b = [6, 0]
x0 = [0, 0]

sol_seidel, trayectoria_seidel = gauss_seidel(A, b, x0, tol=1e-5, max_iter=1000)

print("Solución", sol_seidel)
print("Trayectoria")
for i, x in enumerate(trayectoria_seidel):
    print(f"Iteración {i}: {x}")
```
